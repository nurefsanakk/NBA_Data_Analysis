##### Verilerde hangi kolonlar mevcut ?

```python
#Verilerin içerdiği kolonların isimlerini öğreniyoruz.
def get_columns(group,file_path):
    name = os.path.basename(file_path)
    print(f'{name} dosyasının kolon isimleri : \n{group.columns}')
    print()
    print("*"*100)
    

get_columns(games,games_path)
get_columns(games_details,details_path)
get_columns(players,players_path)
get_columns(ranking,ranking_path)
get_columns(teams,teams_path)
```

##### Kolon veri tipleri neler ?

```python
def get_dtypes(group,file_path):
    name = os.path.basename(file_path)
    print(f'{name} dosyasında bulunan kolonların tipleri : \n{group.dtypes}')
    print()
    print('*'*100)
    print()

get_dtypes(games,games_path)
get_dtypes(games_details,details_path)
get_dtypes(players,players_path)
get_dtypes(ranking,ranking_path)
get_dtypes(teams,teams_path)
```

##### Hangi kolonlarda ne kadar null/boş veri mevcut ?

```python
#verilerde bulunan boş veri sayısını öğreniyoruz.
def get_null(group,file_path):
    name = os.path.basename(file_path)
    print(f'{name} dosyasında bulunan kolonların tipleri : \n{group.isna().sum()}')    
    print()
    print('*'*100)
    print()

get_null(games,games_path)
get_null(games_details,details_path)
get_null(players,players_path)
get_null(ranking,ranking_path)
get_null(teams,teams_path)
    
```

```python
def null_values_table(group): 
    null_values = group.isnull().sum()
    null_values_percent = 100 * group.isnull().sum()/len(group)
    null_values_table = pd.concat([null_values, null_values_percent,group.dtypes], axis=1)
    final_table = null_values_table.rename(
    columns = {0 : 'Eksik Değerler', 1 : '% Değeri', 2: 'Types'})
    return final_table
```

##### Kolonlar içlerinde hangi verileri barındırıyor?

```python
def get_unique_values(group,file_path):
    name = os.path.basename(file_path)
    print(f'{name} dosyası')
    print("-"*60)
    for i in range(len(group.columns)):
        subcolumns = group.columns[i]
        print(f'{subcolumns} Kolonunun Unique değerleri \n{group[subcolumns].value_counts()}')
        print()

get_unique_values(games_details,details_path)
get_unique_values(games,games_path)
get_unique_values(players,players_path)
get_unique_values(ranking,ranking_path)
get_unique_values(teams,teams_path)
```

### Veri Düzenleme
##### Gereksiz Kolonların Silinmesi

```python
games_cleaned = games.copy()
ranking_cleaned = ranking.copy()
teams_cleaned = teams.copy()
games_details_cleaned = games_details.copy()
```

```python
games_cleaned = games_cleaned.drop(columns = ['GAME_STATUS_TEXT', 'TEAM_ID_home','TEAM_ID_away'])
ranking_cleaned = ranking_cleaned.drop(columns = ['LEAGUE_ID','RETURNTOPLAY'])
teams_cleaned = teams_cleaned.drop(columns = ['MAX_YEAR'])
games_details_cleaned = games_details_cleaned.drop(columns = ['NICKNAME','COMMENT','START_POSITION'])
```

##### Kolon İsimlerinin Anlaşılır Hale Getirilmesi

```python
games_cleaned=games_cleaned.rename(columns = {'HOME_TEAM_WINS':'Victory'})
ranking_cleaned=ranking_cleaned.rename(columns = {'G':'Games_Count',
                                                  'W':'Win_Games_Count',
                                                  'L':'Lose_Game_Count'})
```

```python
ranking_cleaned.columns
```

##### Tarih Verilerinin Düzenlenmesi

```python
games_cleaned['GAME_DATE_EST']= pd.to_datetime(games_cleaned['GAME_DATE_EST'])
ranking_cleaned['STANDINGSDATE'] = pd.to_datetime(ranking_cleaned['STANDINGSDATE'])
```

##### Games Details Dosyasında Bulunan MIN değerinin Düzenlenmesi

```python
#Sadece dakika değerini kapsayacak şekilde düzenlenmiş ve kategorik değer numeric hale getirilmiştir.
new = games_details_cleaned["MIN"].str.split(":", n = 1, expand = True)
games_details_cleaned["MIN_m"]= new[0]
games_details_cleaned.drop(columns =["MIN"], inplace = True)
games_details_cleaned["MIN_m"]=pd.to_numeric(games_details_cleaned["MIN_m"])
games_details_cleaned=games_details_cleaned.rename(columns = {'MIN_m':'MIN'})
```

##### Null Değerlerin Temizlenmesi

```python
games_cleaned=games_cleaned.apply(lambda x: x.fillna(x.mean()),axis=0)
```

##### Ev Sahibi ve Deplasman Takımlarının Kazanma-Kaybetme Durumuna Göre Rebound Grafikleri

```python
rebound_win = games_cleaned.groupby(['REB_home', 'Victory'])
rebound_table = rebound_win.size().unstack(fill_value=0)

plt.subplot(1,2,1)
plt.plot(rebound_table[1], color = '#33AB5F',label = 'Rebounds of Home Winner')
plt.plot(rebound_table[0], color = '#BA0001',label = 'Rebounds of Home Loser')
plt.ylabel('Number of Games')
plt.xlabel('Rebounds per Game')
plt.legend()

plt.subplot(1,2,2)
rebound_lose = games_cleaned.groupby(['REB_away','Victory'])
losereb_table = rebound_lose.size().unstack(fill_value=0)

plt.plot(losereb_table[0], color = '#33AB5F',label = 'Rebounds of Away Winner')
plt.plot(losereb_table[1], color = '#BA0001',label = 'Rebounds of Away Loser')

plt.yticks(c='w')
plt.xlim([15,75])
fig = plt.gcf()
fig.set_size_inches(18.5,6)
plt.xlabel('Rebounds per Game')
plt.legend()

```

##### Ev Sahibi ve Deplasman Takımlarının Kazanma-Kaybetme Durumuna Göre Asist Grafikleri

```python
ast_win= games_cleaned.groupby(['AST_home', 'Victory'])
ast_table = ast_win.size().unstack(fill_value=0)

plt.subplot(1,2,1)

plt.plot(ast_table[1], color = '#33AB5F',label = 'Assists of Home Winner')
plt.plot(ast_table[0], color = '#BA0001',label = 'Assists of Home Loser')

plt.xlabel('Assists per Game')
plt.legend()

plt.subplot(1,2,2)
lose_ast_grp = games_cleaned.groupby(['AST_away','Victory'])
lose_ast_table = lose_ast_grp.size().unstack(fill_value=0)

plt.plot(lose_ast_table[0], color = '#33AB5F',label = 'Assists of Away Winner')
plt.plot(lose_ast_table[1], color = '#BA0001',label = 'Assists of Away Loser')

plt.yticks(c='w')
plt.xlabel('Assists per Game')
fig = plt.gcf()
fig.set_size_inches(18.5,6)
plt.legend()
```
